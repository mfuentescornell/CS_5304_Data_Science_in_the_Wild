```python
import pandas as pd
import numpy as np
from unidecode import unidecode
```

```python
import requests

# import urllib library
from urllib.request import urlopen
  
# import json
import json
# store the URL in url as 
# parameter for urlopen
url = "http://api.citybik.es/v2/networks?fields=stations"
  
# store the response of URL
response = urlopen(url)
  
# storing the JSON response 
# from url in data
data_json = json.loads(response.read())
  
# print the json response
print(data_json)
```

```python
# convert json to dataframe
df = pd.json_normalize(data_json['networks'])

# clean text data
df['location.city'] = [unidecode(x) for x in df['location.city']]
df['location.city'] = df['location.city'].str.lower()
df[['location.city', 'location.state']] = df['location.city'].str.split(', ',1, expand=True)

df.head()
```

```python
df_us = df[df['location.country'] == 'US']
df_us
```

```python
city_list = [
    'new york',
    'los angeles',
    'chicago',
    'houston',
    'phoenix',
    'philadelphia',
    'pittsburgh',
    'las vegas',
    'atlanta',
    'austin',
    'san francisco bay area',
    'portland',
    'washington',
    'boston',
]

df_subset = df[df['location.city'].isin(city_list)]
bikeshare_list = df_subset['name'].tolist()
bikeshare_list
```
