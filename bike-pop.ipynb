```python
import pandas as pd
import plotly.express as px
```

```python
bike_pop = pd.read_csv('data\Toplot_bikes_to_pop.csv')
bike_pop = bike_pop[['city','country_x', 'bikes_to_pop']]
bike_pop.rename({'city':'City','country_x':'Country', 'bikes_to_pop':'Bike Docks per Population'}, axis=1, inplace=True)
bike_pop
```

```python
bike_pop.sort_values(by='Bike Docks per Population', ascending=False)
```

```python
top10_cities = bike_pop.sort_values(by='Bike Docks per Population', ascending=False).reset_index(drop=True)[0:10]
top10_cities
```

```python
sorted_df = bike_pop.sort_values(by='Bike Docks per Population', ascending=False)
fig = px.histogram(sorted_df, x='Bike Docks per Population', nbins=200, marginal='box')
fig.show()
```
